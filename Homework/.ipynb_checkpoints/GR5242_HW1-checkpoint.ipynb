## Problem 2

### a)

```python
import numpy as np

def GSAMPLE(mu, sigma, n):
  d = len(mu)
  G = np.random.multivariate_normal(mu, sigma, size = n)
  return(G)
```

### b)

```python
mu = [5, -3]
sigma = np.array([[1, 2], [2, 4]])
print("Real mean:", mu)
print("Real covariance:", sigma)

G = GSAMPLE(mu, sigma, 1000) # draw n=1000 samples
print("Sample mean:", np.mean(G, axis = 0))
print("Sample covariance:", np.cov(G.T))
```

The program approximately reproduce the input parameters.

### c)

```python
import matplotlib.pyplot as plt
mu1 = [1,1]
sigma1 = np.array([[2, 1], [1, 2]])
G1 = GSAMPLE(mu1, sigma1, 100)
plot1 = plt.scatter(x = G1[:,0], y = G1[:,1])
plt.title("Bivariate normal")
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

mu2 = [1,1,1]
sigma2 = np.array([[5,2,0], [2,3,1], [0,1,1]])
G2 = GSAMPLE(mu2, sigma2, 100)
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.scatter(G2[:,0], G2[:,1], G2[:,2])
plt.title("3-dimensional multivariate normal")
plt.show()
```

## Problem 3

### a)

```python
n = 100
N_bins = 10
def hist_plot(n_sample, n_hist, N_bins, ncol, nrow):
  sample = np.zeros([n_hist, n_sample])
  xlab = "x"
  ylab = "Count"
  fig = plt.figure()
  for i in range(n_hist):
    sample[i, :] = np.random.randn(n)
    ax = fig.add_subplot(ncol,nrow,i+1)
    ax.hist(sample[i,:], bins = N_bins)
    ax.set_xlabel(xlab)
    ax.set_ylabel(ylab)
  plt.suptitle("Sample Size =" + str(n) + ", " + str(N_bins) + " bins")
  plt.show()

hist_plot(n, 4, N_bins, 2, 2)
  
```

### b)

```python
n = 100000
hist_plot(n, 4, N_bins, 2, 2)
```

### c)

```python
N_bins = [10, 100, 1000]
for N in N_bins: 
  hist_plot(n, 1, N, 1, 1)
```

### d)

```python
n = 100
N_bins = 1000
hist_plot(n, 1, N_bins, 1, 1)
```

### e)

Generally, the larger the sample size is, the better the result of visualization. As we can see from *a)* and *b)*, the simulation is unstable when sample size is small, thus affects the reliability of visualization. 

In the mean time, the number of bins we select also affect the quality of histograms. When sample size is large and we divide x-axis into only few bins, the estimates become rough and not accurate; when sample size is small but number of bins is large, the histogram becomes several parallel line segments and we can hardly tell what the distribution it is.
